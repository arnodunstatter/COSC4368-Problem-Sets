# Preprocessing the Training and Validation Datasets

## We create functions to turn training and validation data into BERT and Tensorflow format

```python
def convert_data_to_examples(train, test, DATA_COLUMN, LABEL_COLUMN): 
  train_InputExamples = train_data.apply(lambda x: InputExample(guid=None, 
                                                          text_a = x[DATA_COLUMN], 
                                                          text_b = None,
                                                          label = x[LABEL_COLUMN]), axis = 1)

  validation_InputExamples = val_data.apply(lambda x: InputExample(guid=None, 
                                                          text_a = x[DATA_COLUMN], 
                                                          text_b = None,
                                                          label = x[LABEL_COLUMN]), axis = 1)
  
  return train_InputExamples, validation_InputExamples
  
def convert_examples_to_tf_dataset(examples, tokenizer, max_length=128):
    features = [] # -> will hold InputFeatures to be converted later

    for e in examples:
        # Documentation is really strong for this method, so please take a look at it
        input_dict = tokenizer.encode_plus(
            e.text_a,
            add_special_tokens=True,
            max_length=max_length, # truncates if len(s) > max_length
            return_token_type_ids=True,
            return_attention_mask=True,
            pad_to_max_length=True, # pads to the right by default # CHECK THIS for pad_to_max_length
            truncation=True
        )

        input_ids, token_type_ids, attention_mask = (input_dict["input_ids"],
            input_dict["token_type_ids"], input_dict['attention_mask'])

        features.append(
            InputFeatures(
                input_ids=input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids, label=e.label
            )
        )

    def gen():
        for f in features:
            yield (
                {
                    "input_ids": f.input_ids,
                    "attention_mask": f.attention_mask,
                    "token_type_ids": f.token_type_ids,
                },
                f.label,
            )

    return tf.data.Dataset.from_generator(
        gen,
        ({"input_ids": tf.int32, "attention_mask": tf.int32, "token_type_ids": tf.int32}, tf.int64),
        (
            {
                "input_ids": tf.TensorShape([None]),
                "attention_mask": tf.TensorShape([None]),
                "token_type_ids": tf.TensorShape([None]),
            },
            tf.TensorShape([]),
        ),
    )
```

# Testing Different BERTs


## Defining some functions to help simplify the process

```python
from sklearn import metrics

def preprocess_load_compile_fit_evaluate(models_results, pretrained_type, learning_rate=3e-5, epochs=1, epsilon=1e-08, clipnorm=1.0):
  #This will create a model, fine tune it, evaluate it against the test set, and then modify the passed in models_results dictionary to update it
  
  #first we make our tokenizer
  print(f'Downloading {pretrained_type} tokenizer')
  tokenizer = BertTokenizer.from_pretrained('google/bert_uncased_'+pretrained_type)

  #now we preprocess the training data using our defined functions
  maxLength = 128
  train_InputExamples, validation_InputExamples = convert_data_to_examples(train_data, val_data, 'DATA_COLUMN', 'LABEL_COLUMN')
  training_data = convert_examples_to_tf_dataset(list(train_InputExamples), tokenizer, max_length=maxLength)
  training_data = training_data.shuffle(100).batch(32).repeat(2)
  validation_data = convert_examples_to_tf_dataset(list(validation_InputExamples), tokenizer, max_length=maxLength)
  validation_data = validation_data.batch(32)

  #load the pretrained model
  print(f'Downloading {pretrained_type} model')
  model = TFBertForSequenceClassification.from_pretrained('google/bert_uncased_'+pretrained_type, from_pt=True)

  #now we compile the model
  model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate, epsilon=epsilon, clipnorm=clipnorm), 
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
              metrics=[tf.keras.metrics.SparseCategoricalAccuracy('accuracy')])
  
  #now we fine tune the model
  History = model.fit(training_data, epochs=epochs, validation_data=validation_data)
  
  #make predictions on test set, in batches of 50 as per Steve's instructions
  #prepare batches
  reviews = list(test_data["DATA_COLUMN"])
  batches_of_reviews = []
  for i in range(int(len(test_data)/50)):
    batches_of_reviews.append(reviews[i*50:(i+1)*50])
  
  #produce predictions for each batch
  tf_test_all_predictions = []
  for batch in batches_of_reviews:
    tokenized_batch = tokenizer(batch, max_length = maxLength, padding=True, truncation=True, return_tensors='tf')
    tf_test_outputs = model(tokenized_batch)
    #make predictions based on whichever class has highest probability
    tf_test_predictions = tf.nn.softmax(tf_test_outputs[0], axis=-1)
    tf_test_predictions = tf.argmax(tf_test_predictions, axis=1)
    tf_test_all_predictions += list(tf_test_predictions)
  tf_test_all_predictions = np.array(tf_test_all_predictions)

  #evaluate performance
  labels = list(test_data["LABEL_COLUMN"])
  #confusion matrix
  confusion_matrix = metrics.confusion_matrix(labels, tf_test_all_predictions)
  #accuracy
  accuracy = metrics.accuracy_score(labels, tf_test_all_predictions)
  #precision
  precision = metrics.precision_score(labels, tf_test_all_predictions)
  #recall
  recall = metrics.recall_score(labels, tf_test_all_predictions)
  #F1
  f1 = metrics.f1_score(labels, tf_test_all_predictions)

  print(f"{pretrained_type} completed\n")

  models_results[pretrained_type] = {'Train Accuracy':History.history['val_accuracy'][0], 'Test Accuracy':accuracy, 'Precision':precision, 'Recall': recall, 'F1-Score':f1}
  
```

```python
def try_all_types_record_results(learning_rate=3e-5, epochs=1, epsilon=1e-08, clipnorm=1.0):
  model_types = ['L-2_H-128_A-2', 'L-4_H-256_A-4', 'L-4_H-512_A-8', 'L-8_H-512_A-8']
  models_results = {}
  for model in model_types:
    preprocess_load_compile_fit_evaluate(models_results, model, learning_rate, epochs, epsilon, clipnorm)
  results = pd.DataFrame(models_results).transpose()

  setName = f"Set {len(allResults)+1}"
  #add the new data
  allResults[setName] = results
  allParams[setName] = pd.DataFrame({'learning_rate':[learning_rate], 'epochs':[epochs], 'epsilon':[epsilon], 'clipnorm':[clipnorm]})
  saveAll() #save the new data
  print() #whitespace
  display(results)
```

```python
#some functions for saving and loading the allResults dictionary
drive.mount('/content/drive')
problemSet2Path = '/content/drive/MyDrive/1. BS in CS/Semester 6 (Spring 2022)/COSC 4368 - Artificial Intelligence/Problem Sets/Problem Set 2'

def saveAll(loadFirst=True): #setting loadFirst=False will write over old results
  if loadFirst:
    holdResults = deepcopy(allResults) if len(allResults)>0 else {}
    holdParams = deepcopy(allParams) if len(allParams)>0 else {}
    loadAll()
    allResults.update(holdResults)
    allParams.update(holdParams)
  with open(os.path.join(problemSet2Path, 'allResults.pickle'),'wb') as f:
    pickle.dump(allResults,f)
  with open(os.path.join(problemSet2Path, 'allParams.pickle'),'wb') as f:
    pickle.dump(allParams,f)

def loadAll():
  global allResults
  try:
    with open(os.path.join(problemSet2Path, 'allResults.pickle'),'rb') as f:
      allResults = pickle.load(f)
    global allParams
    with open(os.path.join(problemSet2Path, 'allParams.pickle'),'rb') as f:
      allParams = pickle.load(f)
  except FileNotFoundError:
    print("allResults.pickle and/or allParams.pickle was not found and could not be loaded.")
            
```

## Small Grid Search over learning rate and epochs

```python
LEARNING_RATES= [3e-4, 3e-5, 3e-6]
EPOCHS = [1,2]

learningRates_and_epochs = []
for lr in LEARNING_RATES[::-1]:
  for e in EPOCHS[::-1]:
    learningRates_and_epochs.append((lr,e))

loadAll()
for i, v in enumerate(learningRates_and_epochs):
  if len(allResults) <= i:
    lr,e = v
    print(f"Beginning Set {len(allResults)+1} with learning_rate={lr} and epochs={e}")
    try_all_types_record_results(learning_rate=lr, epochs=e)
    print(f"Finished Set {len(allResults)}\n\n\n")
  
```

## Full Bert Model

```python
#This will create a full model, fine tune it, evaluate it against the test set, and then modify the passed in models_results dictionary to update it
models_results = {}
pretrained_type = "bert-base-uncased"
learning_rate=3e-5
epochs=1
epsilon=1e-08
clipnorm=1.0

#first we make our tokenizer
print(f'Downloading {pretrained_type} tokenizer')
tokenizer = BertTokenizer.from_pretrained(pretrained_type)

#now we preprocess the training data using our defined functions
maxLength = 128
train_InputExamples, validation_InputExamples = convert_data_to_examples(train_data, val_data, 'DATA_COLUMN', 'LABEL_COLUMN')
training_data = convert_examples_to_tf_dataset(list(train_InputExamples), tokenizer, max_length=maxLength)
training_data = training_data.shuffle(100).batch(32).repeat(2)
validation_data = convert_examples_to_tf_dataset(list(validation_InputExamples), tokenizer, max_length=maxLength)
validation_data = validation_data.batch(32)

#load the pretrained model
print(f'Downloading {pretrained_type} model')
model = TFBertForSequenceClassification.from_pretrained(pretrained_type, from_pt=True)

#now we compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate, epsilon=epsilon, clipnorm=clipnorm), 
            loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
            metrics=[tf.keras.metrics.SparseCategoricalAccuracy('accuracy')])

#now we fine tune the model
History = model.fit(training_data, epochs=epochs, validation_data=validation_data)

#make predictions on test set, in batches of 50 as per Steve's instructions
#prepare batches
reviews = list(test_data["DATA_COLUMN"])
batches_of_reviews = []
for i in range(int(len(test_data)/50)):
  batches_of_reviews.append(reviews[i*50:(i+1)*50])

#produce predictions for each batch
tf_test_all_predictions = []
for batch in batches_of_reviews:
  tokenized_batch = tokenizer(batch, max_length = maxLength, padding=True, truncation=True, return_tensors='tf')
  tf_test_outputs = model(tokenized_batch)
  #make predictions based on whichever class has highest probability
  tf_test_predictions = tf.nn.softmax(tf_test_outputs[0], axis=-1)
  tf_test_predictions = tf.argmax(tf_test_predictions, axis=1)
  tf_test_all_predictions += list(tf_test_predictions)
tf_test_all_predictions = np.array(tf_test_all_predictions)

#evaluate performance
labels = list(test_data["LABEL_COLUMN"])
#confusion matrix
confusion_matrix = metrics.confusion_matrix(labels, tf_test_all_predictions)
#accuracy
accuracy = metrics.accuracy_score(labels, tf_test_all_predictions)
#precision
precision = metrics.precision_score(labels, tf_test_all_predictions)
#recall
recall = metrics.recall_score(labels, tf_test_all_predictions)
#F1
f1 = metrics.f1_score(labels, tf_test_all_predictions)

print(f"{pretrained_type} completed\n")

#recording results
models_results[pretrained_type] = {'Train Accuracy':History.history['val_accuracy'][0], 'Test Accuracy':accuracy, 'Precision':precision, 'Recall': recall, 'F1-Score':f1}
results = pd.DataFrame(models_results).transpose()
setName = 'Full Bert Model'
allResults[setName] = results
allParams[setName] = pd.DataFrame({'learning_rate':[learning_rate], 'epochs':[epochs], 'epsilon':[epsilon], 'clipnorm':[clipnorm]})
saveAll() #save the new data
print() #whitespace
display(results)
```
